```python
def convert_units(conversion_type, value):
    """
    Converts units based on the conversion type.
    Parameters:
    conversion_type (str): Type of conversion selected by user.
    value (float): Numeric value to convert.
    Returns:
    float: Converted value.
    """

    if conversion_type == "1":
        return value * 3.28084  # meters to feet
    elif conversion_type == "2":
        return value / 3.28084  # feet to meters
    elif conversion_type == "3":
        return value * 2.20462  # kg to pounds
    elif conversion_type == "4":
        return value / 2.20462  # pounds to kg
    elif conversion_type == "5":
        return value * 0.264172  # liters to gallons
    elif conversion_type == "6":
        return value / 0.264172  # gallons to liters
    else:
        return None


def unit_converter():
    print("Select Conversion Type:")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    choice = input("Enter choice (1-6): ")

    try:
        value = float(input("Enter value to convert: "))
        result = convert_units(choice, value)

        if result is None:
            print("Invalid conversion type selected.")
        else:
            print("Converted value:", result)

    except ValueError:
        print("Error: Please enter a valid numeric value.")

unit_converter()
```

```python
def calculate_sum(numbers):
    """
    Returns the sum of numbers in the list.
    Parameters:
    numbers (list): List of numeric values.
    Returns:
    float: Sum of numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Returns the average of numbers in the list.
    Parameters:
    numbers (list): List of numeric values.
    Returns:
    float: Average of numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Returns the maximum number in the list.
    Parameters:
    numbers (list): List of numeric values.
    Returns:
    float: Maximum value.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Returns the minimum number in the list.
    Parameters:
    numbers (list): List of numeric values.
    Returns:
    float: Minimum value.
    """
    return min(numbers)


def math_operations():
    print("\nSelect Operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter choice (1-4): ")

    try:
        numbers = list(map(float, input("Enter numbers separated by space: ").split()))

        if len(numbers) == 0:
            print("Error: List cannot be empty.")
            return

        if choice == "1":
            print("Sum:", calculate_sum(numbers))
        elif choice == "2":
            print("Average:", calculate_average(numbers))
        elif choice == "3":
            print("Maximum:", find_maximum(numbers))
        elif choice == "4":
            print("Minimum:", find_minimum(numbers))
        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Error: Please enter valid numeric values.")



math_operations()
```

```python
def extract_every_other(lst):
    return lst[::2]
```

```python
def get_sublist(lst, start, end):
    return lst[start:end+1]
```

```python
def reverse_list(lst):
    return lst[::-1]
```

```python
def remove_first_last(lst):
    return lst[1:-1]
```

```python
def get_first_n(lst, n):
    return lst[:n]
```

```python
def get_last_n(lst, n):
    return lst[-n:]
```

```python
def flatten(lst):
    """
    Flattens a nested list by one level.
    """
    result = []
    for sublist in lst:
        result.extend(sublist)
    return result
```

```python
def access_nested_element(lst, indices):
    """
    Access element from nested list using list of indices.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element
```

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total
```

```python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result
```

```python
def find_max(lst):
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum
```

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count
```

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result
```

```python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1.
A_inv = np.linalg.inv(A)
print("\nA * A_inv:\n", np.dot(A, A_inv))

# 2.
AB = np.dot(A,B)
BA = np.dot(B,A)
print("\nAB:\n", AB)
print("BA:\n", BA)

# 3.
print("\n(AB)^T:\n", AB.T)
print("B^T A^T:\n", np.dot(B.T, A.T))
```

```python
A = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])

B = np.array([-1,-3,9])

# Using inverse
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("\nUsing inverse:", X)

# Direct solution
X2 = np.linalg.solve(A,B)
print("\nUsing solve function:", X2)
```
